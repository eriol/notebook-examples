# Plot municipatilities example

This example just plot a municipatilities taken from Istat website: https://www.istat.it/it/archivio/222527

```python
import geopandas
import matplotlib.pyplot as plt

# Size of plots.
plt.rcParams['figure.figsize'] = [30, 15]
```

```python
municipalities = "./data/Com01012021_g_WGS84.shp"
```

```python
municipalities = geopandas.read_file(municipalities)
```

```python
municipalities.count()
```

```python
fig = plt.figure(1, figsize=(40,20)) 
ax = fig.add_subplot()
ax.axis('off')

municipalities.boundary.plot(ax=ax)
```
